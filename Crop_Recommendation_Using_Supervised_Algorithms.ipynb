```python
#boxplot_using_for_loop
plt.figure(figsize=(20,15))
plotnumber=1
for col in df.columns:
  if plotnumber<=8:
    plt.subplot(5,5,plotnumber)
    sns.boxplot(df[col])
    plt.xlabel(col,fontsize=15)
  plotnumber+=1
plt.tight_layout()
plt.show()
```

```python
#minimum_amount_of_rainfall
r=min(df['Rainfall'])
r
```

```python
#finding_the_crop_recommended_in_minimum_amount_of_rainfall
N=df[df['Rainfall']==r]
N[['Rainfall','Crop']]
```

```python
#finding_the_crop_recommended_in_minimum_humidity_and_temperature
b=min(df['Humidity'])
c=min(df['Temperature'])
L=df[(df['Humidity']==b) | (df['Temperature']==c)]
L[['Temperature','Humidity','Crop']]
```

```python
#Label_Encoding
from sklearn.preprocessing import LabelEncoder
a=LabelEncoder()
columns=df.select_dtypes(include=['object']).columns
for col in columns:
  df[col]=a.fit_transform(df[col])
df
```

```python
#heatmap
plt.figure(figsize=(20,15))
corr=df.corr()
sns.heatmap(corr,annot=True,fmt='.2g',linewidth=1)
plt.tight_layout()
plt.show()
```

```python
x=df.drop('Crop',axis=1)
y=df['Crop']
```

```python
#training_and_testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
x_train
```

```python
#support_vector_machine
from sklearn.svm import SVC
clf=SVC(kernel='rbf')
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
#Testing_accuracy_score
accuracy_score(y_pred,y_test)
```

```python
y_pred1=clf.predict(x_train)
```

```python
#Training_accuracy_score
accuracy_score(y_pred1,y_train)
```

```python
#DecisionTree_Algorithm
from sklearn.tree import DecisionTreeClassifier
clf1=DecisionTreeClassifier(criterion='entropy')
clf1.fit(x_train,y_train)
```

```python
y_pred2=clf1.predict(x_test)
```

```python
#Testing_accuracy_score
accuracy_score(y_pred2,y_test)
```

```python
Y_Pred=clf1.predict(x_train)
```

```python
#Training_accuracy_score
accuracy_score(Y_Pred,y_train)
```

```python
#Random_Forest_algorithm
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(random_state=40,n_jobs=-1,max_depth=5,n_estimators=80,oob_score=True)
rf.fit(x_train,y_train)
```

```python
Y_pred=rf.predict(x_test)
```

```python
#Testing_accuracy_score
accuracy_score(Y_pred,y_test)
```

```python
Y_PRed=rf.predict(x_train)
```

```python
#Training_accuracy_score
accuracy_score(Y_PRed,y_train)
```
